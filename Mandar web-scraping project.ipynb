```python
url='https://www.flipkart.com/apple-iphone-15-pro-max-natural-titanium-256-gb/p/itm9d44960698192?pid=MOBGTAGQH6K7YMJU&lid=LSTMOBGTAGQH6K7YMJU74VM64&marketplace=FLIPKART&q=iphone15+pro+max&store=tyy%2F4io&srno=s_1_1&otracker=AS_QueryStore_OrganicAutoSuggest_1_8_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_8_na_na_ps&fm=neo%2Fmerchandising&iid=d4580bec-41e9-4380-b2a2-7f57990595f4.MOBGTAGQH6K7YMJU.SEARCH&ppt=browse&ppn=browse&ssid=47448get5s0000001702808701799&qH=a799493a392de1ae'
```

#### The above is the url we are going to fetch for scrapping the data.

```python
page=requests.get(url)
page
```

##### with the help of requests if we get the response of 503 it means that we can scrap that page.

```python
data=page.text
```

```python
soup=bs(data,"html.parser")
```

```python
print(soup.prettify())
```

##### Beautiful Soup then parses the document using the best available parser. The prettify() method will turn a Beautiful Soup parse tree into a nicely formatted Unicode string,with a separate line for each tag and each string

```python
title=soup.title
print(title.text)
```

```python
price=soup.find("div",class_="_30jeq3 _16Jk6d")
print(price.text)
```

```python
rating=soup.find_all("span")
rating
```

##### we have find all cusomers name from HTML page

```python
names=soup.find_all("p",class_="_2sc7ZR _2V5EHH")
```

```python
names
```

```python
cust_name=[]
for i in range(0,len(names)):
    cust_name.append(names[i].text)
cust_name
```

##### now we have find all the comments given by the customers

```python
comments=soup.find_all("p",class_="_2-N8zT")
comments
```

```python
cust_comment=[]
for i in range(0,len(comments)):
    cust_comment.append(comments[i].text)
cust_comment
print(len(cust_comment))
```

##### now next we have find the rateings given by customers

```python
stars=soup.find_all("div",class_="_3LWZlK _1BLPMq")
stars
```

```python
cust_star=[]
for i in range(0,len(stars)):
    cust_star.append(stars[i].text)
cust_star
```

##### now will find the reviews given from the customers 

```python
reviews=soup.find_all("div",class_="t-ZTKy")
reviews
```

```python
cust_rev=[]
for i in range(0,len(reviews)):
    cust_rev.append(reviews[i].text)
cust_rev
```

##### likes and dislike from the customers

```python
like=soup.find_all("span",class_="_3c3Px5")
like
```

```python
cust_like=[]
for i in range(0,len(like)):
    cust_like.append(like[i].text)
print(cust_like,end=" ")
```

```python
import pandas as pd
```

```python
df=pd.DataFrame()
df["customer_name"]=cust_name
df["comments"]=cust_comment
df["reviews"]=cust_rev
df
```

```python
df.to_csv("data.csv",index=False)
```
